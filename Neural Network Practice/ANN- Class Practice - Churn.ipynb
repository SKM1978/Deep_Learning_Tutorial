# Data Cleaning / Preprocessing

# Data Creation - Modelling

```python
df['Geography'].unique()
```

```python
df['Gender']=df['Gender'].map({'Female':1, 'Male':0}).astype('int')
df['Geography']=df['Geography'].map({'France':0, 'Spain':1, 'Germany':2}).astype('int')
```

```python
numerical_variables+categorical_variables
```

```python
churn=df.loc[:,numerical_variables+categorical_variables]
```

## Split variables

```python
X=churn.drop(columns='Exited')
y=churn.loc[:,'Exited']
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)
```

```python
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train.shape,X_test.shape
```

# Model Creation

## Stats Based Approach

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=5)
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
```

# Model Evaluation

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

# Model Save / Load

```python
filename="./rf_simple_classifier.pkl"
import pickle
pickle.dump(rf,open(filename,"wb"))
```

```python
rf2=pickle.load(open(filename,"rb"))
```

```python
print(classification_report(y_test,rf2.predict(X_test)))
```
